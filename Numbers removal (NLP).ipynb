```python
#Importing all the necessary libraries....
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import pandas as pd
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
import nltk
```

# Prerequisite work 

```python
#Reading the csv file in jupyter.
df=pd.read_csv('SPAM text message 20170820 - Data - SPAM text message 20170820 - Data.csv')
```

```python
#To display the dataframe.
df
```

```python
#Creating a function to remove all the stopwords in a sentence.
def datacleaning(x):
    y=nltk.word_tokenize(x)
    temp=[]
    for i in y:
      if i in stopwords.words("english"): 
         pass
      else:
         temp.append(i)
    my_new_string=' '.join(temp)
    return (my_new_string)
```

```python
#Using the function on the column Message and storing the text in a new column called Clean_message.
df['Clean_message']=df['Message'].apply(datacleaning)
```

```python
#display the dataframe
df
```

```python
#Creating a function to lemmatize words in a sentence.
def my_lemmatizer(a):
    c=nltk.word_tokenize(a)
    my_list=[]
    for i in c:
        my_list.append(lemmatizer.lemmatize(i))
        my_new_string=' '.join(my_list)
    return my_new_string
```

```python
#Using the function to lemmatize the text in every row of the clean_message and saving the lemmatized text in a ne column 
#called Lemmatized_message.
df['Lemmatized_message']=df['Clean_message'].apply(my_lemmatizer)
```

```python
#Display the dataframe.
df
```

```python
#Creating a function to count the length of text.
def Text_count(x):
    y=nltk.word_tokenize(x)
    temp=[]
    for i in y: 
         temp.append(i)
    d=' '.join(temp)
    return(len(d.split()))
```

```python
#counting the length of text in clean message using the function and saving it in a new colomn called length_text.
df['Lenghth_text']=df['Clean_message'].apply(Text_count)
```

```python
#Display the dataframe
df
```

```python
#Removing the numbers from the text.
def rem_num(x):
    temp=[]
    for i in x:
      if i.isdigit(): 
         pass
      else:
        temp.append(i)
    hi=''.join(temp)
    return hi
```

```python
#Testing the function in a dummy text to see the impact.
rem_num('Hi i am 10 rasesh 20 and 30')
```

```python
#applying the function created to remove all the numbers form clean message to extract only the text and saving it in a new column
#called Rem_num.
df['Rem_num']=df['Clean_message'].apply(rem_num)
```

```python
#Display the dataframe
df
```

```python
#After removing the numbers from clean message and saving it in rem_num column, we finally vectorize our column rem_num
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
```

```python
#Fitting the text/string/column(Rem_num) in the vector.
vectorizer.fit(df['Rem_num'])
```

```python
#Getting the vocab out.
vectorizer.get_feature_names_out()
```

```python
#Getting the the first 100 vocabs out:
vectorizer.get_feature_names_out()[0:100]
```

```python
#Transforming the text/string/column(Rem_num) into vector
vector=vectorizer.transform(df['Rem_num'])
```

```python
#To display the sparse matrix.
vector
```

```python
#Converting the sparse matrix to array form and saving it in a a variable called x.
x=vector.toarray()
```
